## Handle NaNs


```python
df_cleaned = vd_fpv_df.dropna(subset=['Content'])
```

```python
df_cleaned.iloc[:20]
```

## Replace '\n' with ' '

```python
df_cleaned.loc[:, 'Content'] = df_cleaned['Content'].str.replace('\n', ' ', regex=False)
```

```python
df_cleaned.iloc[:20]
```

## Combine messages

```python
df_cleaned['Group'] = (df_cleaned['AuthorID'] != df_cleaned['AuthorID'].shift()).cumsum()

df_combined = df_cleaned.groupby(['Group', 'AuthorID'], as_index=False).agg({
    'Author': 'first',       # Keep the first Author
    'Content': '. '.join,   # Combine 'Content'
    'Date': 'first',        # Keep the first Date in each group
})
df_combined = df_combined.drop(columns=['Group'])
```

## Deal with emojis


```python
import emoji

def remove_emojis(text):
    return emoji.replace_emoji(text, "")


def demojize_reaction(text):
    if isinstance(text, str):
        return emoji.demojize(text) 
    else:
        return text
```

### Remove emojis

```python
df_cleaned_no_emojis = df_combined.copy()
df_cleaned_no_emojis['Content'] = df_cleaned_no_emojis['Content'].apply(remove_emojis)
df_cleaned_no_emojis.iloc[:10]
```

Entries with empty content will be removed later

### Convert emojis to text description

```python
df_cleaned_demojized = df_combined.copy()
df_cleaned_demojized['Content'] = df_cleaned_demojized['Content'].apply(lambda text: emoji.demojize(text))
```

```python
df_cleaned_demojized.iloc[:10]
```

## Filter relevant content

```python
# Define a filtering function to exclude non-technical content
def filter_relevant(content):
    if isinstance(content, str) and len(content) > 20:  # Exclude very short messages like "Hi"
        return True
    return False

data_filtered = df_cleaned_demojized[df_cleaned_demojized['Content'].apply(filter_relevant)]
```

## Group Messages into Conversations

```python
data_filtered['Date'] = pd.to_datetime(data_filtered['Date'])
data = data_filtered.sort_values(by='Date')

# Example logic to create threads based on a time window
threads = []
current_thread = []
for _, row in data.iterrows():
    if current_thread and (row['Date'] - current_thread[-1]['Date']).total_seconds() > 1500:
        threads.append(current_thread)
        current_thread = []
    current_thread.append(row)
if current_thread:
    threads.append(current_thread)
```

```python
print(f"threads: {len(threads)}\n{threads[4]}")
```

## Extract questions and answers

```python
qa_pairs = []

for thread in threads:
    question = None
    answer = []
    for message in thread:
        if '?' in message['Content']:  # Identify a question
            if question:  # If we already have a question, finalize the previous Q&A pair
                qa_pairs.append({'Question': question, 'Answer': ' '.join(answer)})
                answer = []
            question = message['Content']
        else:
            answer.append(message['Content'])

    # Append the last Q&A pair
    if question:
        qa_pairs.append({'Question': question, 'Answer': ' '.join(answer)})
```

```python
qa_pairs
```

```python
valid_qa_pairs = []
for p in qa_pairs:
    if p['Question'] != '' and p['Answer'] != '':
        valid_qa_pairs.append(p)
```

```python
qa_df = pd.DataFrame(valid_qa_pairs)
qa_df.to_csv('../data/converted/fpv_qa_formatted.csv', index=False)
```
