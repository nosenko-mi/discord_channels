## Handle NaNs


```python
df_cleaned = vd_fpv_df.dropna(subset=['Content'])
```

```python
df_cleaned.iloc[:20]
```

## Replace '\n' with ' '

```python
df_cleaned.loc[:, 'Content'] = df_cleaned['Content'].str.replace('\n', ' ', regex=False)
```

```python
df_cleaned.iloc[:20]
```

## Combine messages

```python
df_cleaned['Group'] = (df_cleaned['AuthorID'] != df_cleaned['AuthorID'].shift()).cumsum()

df_combined = df_cleaned.groupby(['Group', 'AuthorID'], as_index=False).agg({
    'Content': '. '.join,   # Combine 'Content'
    'Author': 'first',       # Keep the first Author
    'Date': 'first',        # Keep the first Date in each group
    'Reactions': 'last',   # Keep the first Reactions (or use another method)
})
df_combined = df_combined.drop(columns=['Group'])
```

```python
print(df_combined.iloc[15])
```

```python
print(df_combined['Content'].iloc[15])
```

## Deal with emojis


```python
import emoji

def remove_emojis(text):
    return emoji.replace_emoji(text, "")


def demojize_reaction(text):
    if isinstance(text, str):
        return emoji.demojize(text) 
    else:
        return text
```

### Remove emojis

```python
df_cleaned_no_emojis = df_combined.copy()
```

```python
df_cleaned_no_emojis['Content'] = df_cleaned_no_emojis['Content'].apply(remove_emojis)
```

```python
df_cleaned_no_emojis.iloc[:20]
```

### Convert emojis to text description

```python
df_cleaned_demojized = df_combined.copy()
df_cleaned_demojized['Content'] = df_cleaned_demojized['Content'].apply(lambda text: emoji.demojize(text))
```

```python
df_cleaned_demojized.iloc[:20]
```

```python
df_cleaned_demojized['Reactions'] = df_cleaned_demojized['Reactions'].apply(demojize_reaction)
```

```python
df_cleaned_demojized.iloc[:20]
```

## Save dataframes

```python
df_cleaned_demojized.to_csv('../data/converted/vd-fpv-demojized.csv', index=False)
df_cleaned_no_emojis.to_csv('../data/converted/vd-fpv-no-emojis.csv', index=False)
```
